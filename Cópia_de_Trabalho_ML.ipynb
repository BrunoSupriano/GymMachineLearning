```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv('gym_members_exercise_tracking.csv', index_col=False)
```

```python
dataset
```

```python
dataset.info()
```

```python
dataset.columns
```

```python
from sklearn import preprocessing
```

```python
scaler = preprocessing.MinMaxScaler()
```

```python
df = dataset
```

```python
colunas_para_normalizar = [
    'Weight (kg)', 'Max_BPM', 'Avg_BPM',
    'Resting_BPM', 'Calories_Burned', 'Fat_Percentage', 'BMI'
]
```

```python
df[colunas_para_normalizar] = scaler.fit_transform(df[colunas_para_normalizar])
```

```python
df
```

```python
columns = df.select_dtypes(include=['object', 'bool']).columns
```

```python
columns
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
ordinal = OrdinalEncoder()
```

```python
values_cat = ordinal.fit_transform(df[columns])
```

```python
df[columns]=values_cat
```

```python
df
```

```python
x = df.drop(['Workout_Type'], axis=1)
```

```python
x
```

```python
y = df['Workout_Type']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.30, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model_regression = LinearRegression()
```

```python
model_regression.fit(x_train, y_train)
```

```python
model_regression.coef_
```

```python
y_predict = model_regression.predict(x_test)
```

```python
y_predict
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
mean_absolute_error(y_test, y_predict)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test, y_predict)
```
